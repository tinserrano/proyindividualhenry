Necesitamos desanidar las columnas Genres, Belongs_to_Collection y Spoken_Languages

Desanidamos Belongs_to_Collection

```python
import ast

# Función para evaluar la cadena de texto y obtener el nombre de la colección
def extract_collection_name(collection):
    try:
        collection_dict = ast.literal_eval(collection)
        if isinstance(collection_dict, dict):
            return collection_dict.get('name', pd.NA)
    except (ValueError, SyntaxError):
        pass
    return pd.NA

# Aplicar la función con apply y crear una nueva columna "collection_name"
df['collection_name'] = df['belongs_to_collection'].apply(extract_collection_name)


def extract_collection_id(collection):
    try:
        collection_dict = ast.literal_eval(collection)
        if isinstance(collection_dict, dict):
            return collection_dict.get('id', pd.NA)
    except (ValueError, SyntaxError):
        pass
    return pd.NA

# Aplicar la función con apply y crear una nueva columna "collection_name"
df['collection_name_id'] = df['belongs_to_collection'].apply(extract_collection_id)
```

Desanidamos spoken_languages

```python
df["spoken_languages"]
```

```python
# Función para extraer el nombre del idioma o devolver NaN si la lista está vacía o no es válida
def extract_language_ISO(language_list):
    try:
        if isinstance(language_list, str):
            language_list = ast.literal_eval(language_list)

        if isinstance(language_list, list) and len(language_list) > 0:
            language_dict = language_list[0]
            if isinstance(language_dict, dict):
                return language_dict.get('iso_639_1', pd.NA)
    except (ValueError, SyntaxError):
        pass
    return pd.NA

# Aplicar la función con apply y crear una nueva columna "language_name"
df['language_name_ISO'] = df['spoken_languages'].apply(extract_language_ISO)



# Función para extraer el nombre del idioma o devolver NaN si la lista está vacía o no es válida
def extract_language_name(language_list):
    try:
        if isinstance(language_list, str):
            language_list = ast.literal_eval(language_list)

        if isinstance(language_list, list) and len(language_list) > 0:
            language_dict = language_list[0]
            if isinstance(language_dict, dict):
                return language_dict.get('name', pd.NA)
    except (ValueError, SyntaxError):
        pass
    return pd.NA

# Aplicar la función con apply y crear una nueva columna "language_name"
df['language_name'] = df['spoken_languages'].apply(extract_language_name)

```

Desanidamos Genres

```python
df["genres"]
```

```python
# Función para obtener los IDs separados por coma
def extract_genre_ids(genre_list):
    try:
        if isinstance(genre_list, str):
            genre_list = ast.literal_eval(genre_list)
            
        genre_ids = [genre['id'] for genre in genre_list]
        return ','.join(map(str, genre_ids))
    except (ValueError, SyntaxError):
        return pd.NA

# Aplicar la función con apply y crear una nueva columna "genre_ids"
df['genre_ids'] = df['genres'].apply(extract_genre_ids)


# Función para obtener los names separados por coma
def extract_genre_names(genre_list):
    try:
        if isinstance(genre_list, str):
            genre_list = ast.literal_eval(genre_list)
            
        genre_ids = [genre['name'] for genre in genre_list]
        return ','.join(map(str, genre_ids))
    except (ValueError, SyntaxError):
        return pd.NA

# Aplicar la función con apply y crear una nueva columna "genre_ids"
df['genre_name'] = df['genres'].apply(extract_genre_names)
```

Desanidamos production_companies

```python
df["production_companies"]
```

```python
def extract_production_companies(companies_list):
    try:
        companies_list = ast.literal_eval(companies_list)
        if isinstance(companies_list, list):
            return ', '.join([company['name'] for company in companies_list])
    except (ValueError, SyntaxError):
        pass
    return pd.NA

# Aplicar la función con apply y crear una nueva columna "production_companies_names"
df['production_companies_names'] = df['production_companies'].apply(extract_production_companies)

def extract_production_companies(companies_list):
    try:
        companies_list = ast.literal_eval(companies_list)
        if isinstance(companies_list, list):
            return ', '.join([company['name'] for company in companies_list])
    except (ValueError, SyntaxError):
        pass
    return pd.NA

# Aplicar la función con apply y crear una nueva columna "production_companies_names"
df['production_companies_names'] = df['production_companies'].apply(extract_production_companies)


def extract_production_companies_ids(companies_list):
    try:
        companies_list = ast.literal_eval(companies_list)
        if isinstance(companies_list, list):
            return ', '.join([str(company['id']) for company in companies_list])
    except (ValueError, SyntaxError):
        pass
    return pd.NA

# Aplicar la función con apply y crear una nueva columna "production_companies_ids"
df['production_companies_ids'] = df['production_companies'].apply(extract_production_companies_ids)
```

Desanidamos	production_countries

```python
df["production_countries"]
```

```python
def extract_production_countries(countries_list):
    try:
        countries_list = ast.literal_eval(countries_list)
        if isinstance(countries_list, list):
            iso_list = [country['iso_3166_1'] for country in countries_list]
            name_list = [country['name'] for country in countries_list]
            return ', '.join(iso_list), ', '.join(name_list)
    except (ValueError, SyntaxError):
        pass
    return pd.NA, pd.NA

# Aplicar la función con apply y crear dos nuevas columnas "iso_3166_1" y "country_name"
df[['production_country_id', 'production_country_name']] = df['production_countries'].apply(extract_production_countries).apply(pd.Series)
```

```python

# Función para corregir o marcar valores incorrectos en la columna 'release_date'
def fix_release_date(date_str):
    try:
        # Intentar convertir la fecha a datetime con el formato '%Y %m %d'
        return pd.to_datetime(date_str, format='%Y %m %d')
    except ValueError:
        # Si no se puede convertir, devuelve None para indicar que debe revisarse manualmente
        return None

# Utilizar apply y assign para hacer la asignación sin advertencias
df = df.assign(release_date=df['release_date'].apply(fix_release_date))
```

```python
df["year"] = df["release_date"].dt.year
```

```python
df
```

```python
def calculate_roi(revenue, budget):
    try:
        revenue = pd.to_numeric(revenue)
        budget = pd.to_numeric(budget)

        if pd.notna(revenue) and pd.notna(budget) and budget != 0:
            return revenue / budget
        else:
            return 0
    except (ValueError, TypeError):
        return 0

# Aplicar la función con apply y crear una nueva columna 'return'
df['return'] = df.apply(lambda row: calculate_roi(row['revenue'], row['budget']), axis=1)
```
